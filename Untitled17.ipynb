```python
#preprocessing
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize
```

```python
target=[]
images=[]
flat_data=[]
```

```python
DATADIR='/content/drive/MyDrive/tomato3'
CATAGORIES=['tomato bacterial spot','tomato early blight','tomato healthy','tomato late blight','tomato leaf mold','tomato mosaic virus','tomato septoria leaf','tomato target spot','tomato two spotted spider mite','tomato yellow leaf curl virus']
```

```python
for catagory in CATAGORIES:
  class_num=CATAGORIES.index(catagory)
  path=os.path.join(DATADIR,catagory)
  for img in os.listdir(path):
    img_array=imread(os.path.join(path,img))
    img_resized=resize(img_array,(150,150,3))
    flat_data.append(img_resized.flatten())
    images.append(img_resized)
    target.append(class_num)
```

```python
flat_data=np.array(flat_data)
target=np.array(target)
images=np.array(images)
```

```python
#splitting the data into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(flat_data,target,test_size=0.3,random_state=109)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid=[
            {'C':[1,10,100,1000],"kernel":['linear']},
            {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']},
            
]
svc=svm.SVC(probability=True)
clf=GridSearchCV(svc,param_grid)
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_pred,y_test)
```
